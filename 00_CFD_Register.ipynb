```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_excel("./0.DataSource/CfD Register 3-Jul-2024.xlsx",sheet_name='Register')
# https://cfd.lowcarboncontracts.uk/cfd-register/ ## Prices are in 2024.
```

```python
DataStatus = pd.read_csv("./0.DataSource/cfd_contract_portfolio_status.csv")
```

```python
DataStatus.columns
```

```python
DataStatus.head()
```

```python
SelCols1 = ["Unique Identifier","Name of CFD Unit","Start Date","CFD Register Entry was Last Updated","Allocation round","Technology type","Initial Installed Capacity Estimate for CFD Unit, MW","Current strike price"]
Data = Data[SelCols1]
```

```python
SelCols2 = ["CFD_ID","Status"]
DataStatus = DataStatus[SelCols2]
```

```python
Data['CombinedPrice'] = Data['Initial Installed Capacity Estimate for CFD Unit, MW'] * Data['Current strike price']
```

```python
Data.head()
```

```python
DataStatus.head()
```

```python
MergedDF = pd.merge(left=Data, right=DataStatus, how='left', left_on='Unique Identifier', right_on='CFD_ID')
DropColumns = ['CFD_ID']
MergedDF.drop(columns=DropColumns,inplace=True)
```

```python
MergedDF["Allocation round"].unique()
```

```python
MergedDF.to_csv("./0.DataSource/CFD_Contracts_MERGED.csv",index=False)
```

```python
MergedDF
```

```python
PvtTable = pd.pivot_table(MergedDF,values=["CombinedPrice","Initial Installed Capacity Estimate for CFD Unit, MW"],index=["Technology type","Allocation round","Status"],aggfunc='sum')
```

```python
PvtTable["VolumeWeighted,£/MWh"] = PvtTable["CombinedPrice"] / PvtTable["Initial Installed Capacity Estimate for CFD Unit, MW"]
PvtTable.drop(columns=["CombinedPrice"],inplace=True)
```
